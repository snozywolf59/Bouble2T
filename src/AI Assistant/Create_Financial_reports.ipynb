### Bao cao gia thanh san pham

```python
# List of inputs
'''
- ma_sp: array[string]
- ten_sp: array[string]
- dv_tinh: array[string]
- sl_sx: np.array[int]
- chi_phi_nvl: np.array[float]
- chi_phi_nhan_cong: np.array[float]
- chi_phi_chung: np.array[float]
- gia_ban_dv: np.array[float] = chi_phi_nvl + chi_phi_nhan_cong + chi_phi_chung
- gia_ban_tong: np.array[float] = gia_ban_dv * sl_sx

'''

def GTSP(ma_sp, ten_sp, dv_tinh, sl_sx, chi_phi_nvl, chi_phi_nhan_cong, chi_phi_chung):
    gia_ban_dv = chi_phi_nvl + chi_phi_nhan_cong + chi_phi_chung
    gia_ban_tong = np.dot(gia_ban_dv, sl_sx)
    df = pd.DataFrame({'Mã sản phẩm': ma_sp, 'Tên sản phẩm': ten_sp, 'Đơn vị tính': dv_tinh, 'Số lượng sản xuất': sl_sx, 'Chi phí nguyên vật liệu': chi_phi_nvl, 'Chi phí nhân công': chi_phi_nhan_cong, 'Chi phí chung': chi_phi_chung, 'Giá bán đơn vị': gia_ban_dv, 'Giá bán tổng': gia_ban_tong})
    df.to_excel('test.xlsx', sheet_name='sheet1', index=False, startrow=4, startcol=1)
    return df
```

```python
# test block
ma_sp = ['SP01', 'SP02', 'SP03']
ten_sp = ['Sản phẩm 1', 'Sản phẩm 2', 'Sản phẩm 3']
dv_tinh = ['Cái', 'Cái', 'Cái']
# the remaining inputs are random in type of numpy arrays
sl_sx = np.random.randint(100, 200, 3)
chi_phi_nvl = np.random.randint(100, 200, 3)
chi_phi_nhan_cong = np.random.randint(100, 200, 3)
chi_phi_chung = np.random.randint(100, 200, 3)
a = GTSP(ma_sp, ten_sp, dv_tinh, sl_sx, chi_phi_nvl, chi_phi_nhan_cong, chi_phi_chung)
```

### Bao cao cong no

```python
# List of inputs
'''
- ma-kh
- ten-kh
- dia-chi
- ma-so-thue
- no_dau_ky
- co_dau_ky
- no_phat_sinh
- co_phat_sinh
- no_cuoi_ky = max(no_dau_ky + no_phat_sinh - co_dau_ky - co_phat_sinh, 0)
- co_cuoi_ky = max(co_dau_ky + co_phat_sinh - no_dau_ky - no_phat_sinh, 0)

'''

def BCCN(ma_kh, ten_kh, dia_chi, ma_so_thue, no_dau_ky, co_dau_ky, no_phat_sinh, co_phat_sinh):
    no_cuoi_ky = np.maximum(no_dau_ky + no_phat_sinh - co_dau_ky - co_phat_sinh, 0)
    co_cuoi_ky = np.maximum(co_dau_ky + co_phat_sinh - no_dau_ky - no_phat_sinh, 0)
    df = pd.DataFrame({'Mã khách hàng': ma_kh, 'Tên khách hàng': ten_kh, 'Địa chỉ': dia_chi, 'Mã số thuế': ma_so_thue, 'Nợ đầu kỳ': no_dau_ky, 'Có đầu kỳ': co_dau_ky, 'Nợ phát sinh': no_phat_sinh, 'Có phát sinh': co_phat_sinh, 'Nợ cuối kỳ': no_cuoi_ky, 'Có cuối kỳ': co_cuoi_ky})
    df.to_excel('test.xlsx', sheet_name='Sheet4', index=False, startrow=4, startcol=1)
    return df

# test block
ma_kh = ['KH01', 'KH02', 'KH03']
ten_kh = ['Khách hàng 1', 'Khách hàng 2', 'Khách hàng 3']
dia_chi = ['Địa chỉ 1', 'Địa chỉ 2', 'Địa chỉ 3']
ma_so_thue = ['MST01', 'MST02', 'MST03']
no_dau_ky = np.random.randint(100, 200, 3)
co_dau_ky = np.random.randint(100, 200, 3)
no_phat_sinh = np.random.randint(100, 200, 3)
co_phat_sinh = np.random.randint(100, 200, 3)
a = BCCN(ma_kh, ten_kh, dia_chi, ma_so_thue, no_dau_ky, co_dau_ky, no_phat_sinh, co_phat_sinh)
a
```
