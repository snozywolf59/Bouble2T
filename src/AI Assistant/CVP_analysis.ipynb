```python
# Blocks for formulae

def cvp_analysis(gia_ban_sp, bien_phi_per_sp, doanh_thu, tong_chi_phi, tong_chi_phi_co_dinh, tong_bien_phi, loi_nhuan_muc_tieu, gia_ban_sp_muc_tieu):

    loi_nhuan_gop_per_sp = gia_ban_sp - bien_phi_per_sp
    loi_nhuan_gop_over_doanh_thu = loi_nhuan_gop_per_sp / gia_ban_sp
    tong_loi_nhuan_gop = doanh_thu - tong_bien_phi

    loi_nhuan = doanh_thu - tong_chi_phi

    diem_hoa_von = tong_chi_phi_co_dinh / loi_nhuan_gop_per_sp
    SL_SP_can_ban_hoa_von = (tong_chi_phi_co_dinh + loi_nhuan_muc_tieu) / loi_nhuan_gop_per_sp
    
    bien_do_an_toan_tuyet_doi = gia_ban_sp - gia_ban_sp_muc_tieu
    bien_do_an_toan_tuong_doi = bien_do_an_toan_tuyet_doi / gia_ban_sp
    
    return loi_nhuan_gop_per_sp, loi_nhuan_gop_over_doanh_thu, tong_loi_nhuan_gop, loi_nhuan, diem_hoa_von, SL_SP_can_ban_hoa_von, bien_do_an_toan_tuyet_doi, bien_do_an_toan_tuong_doi

def cvp_analysis_plot(gia_ban_sp, bien_phi_per_sp, doanh_thu, chi_phi, tong_chi_phi_co_dinh, loi_nhuan_muc_tieu, gia_ban_sp_muc_tieu):
    plt.figure(figsize=(10, 10))
    loi_nhuan_gop_per_sp, loi_nhuan_gop_over_doanh_thu, loi_nhuan, diem_hoa_von, SL_SP_can_ban_hoa_von, bien_do_an_toan_tuyet_doi, bien_do_an_toan_tuong_doi = cvp_analysis(gia_ban_sp, bien_phi_per_sp, doanh_thu, chi_phi, tong_chi_phi_co_dinh, loi_nhuan_muc_tieu, gia_ban_sp_muc_tieu)
    loi_nhuan_gop_per_sp = np.linspace(loi_nhuan_gop_per_sp, loi_nhuan_gop_per_sp, 100)
    loi_nhuan_gop_over_doanh_thu = np.linspace(loi_nhuan_gop_over_doanh_thu, loi_nhuan_gop_over_doanh_thu, 100)
    diem_hoa_von = np.linspace(diem_hoa_von, diem_hoa_von, 100)
    SL_SP_can_ban_hoa_von = np.linspace(SL_SP_can_ban_hoa_von, SL_SP_can_ban_hoa_von, 100)
    bien_do_an_toan_tuyet_doi = np.linspace(bien_do_an_toan_tuyet_doi, bien_do_an_toan_tuyet_doi, 100)
    bien_do_an_toan_tuong_doi = np.linspace(bien_do_an_toan_tuong_doi, bien_do_an_toan_tuong_doi, 100)
    plt.plot(loi_nhuan_gop_per_sp, np.linspace(0, 100, 100), label='loi_nhuan_gop_per_sp')
    plt.plot(loi_nhuan_gop_over_doanh_thu, np.linspace(0, 100, 100), label='loi_nhuan_gop_over_doanh_thu')
    plt.plot(diem_hoa_von, np.linspace(0, 100, 100), label='diem_hoa_von')
    plt.plot(SL_SP_can_ban_hoa_von, np.linspace(0, 100, 100), label='SL_SP_can_ban_hoa_von')
    plt.plot(bien_do_an_toan_tuyet_doi, np.linspace(0, 100, 100), label='bien_do_an_toan_tuyet_doi')
    plt.plot(bien_do_an_toan_tuong_doi, np.linspace(0, 100, 100), label='bien_do_an_toan_tuong_doi')
    plt.legend()
    plt.show()
```

```python
# Thay doi chi phi co dinh va doanh thu

def optimize_chiphicodinh(gia_ban_sp, doanh_thu, so_luong_sp, so_luong_sp_moi, tong_chi_phi_co_dinh, tong_chi_phi_co_dinh_moi, tong_bien_phi):
    # d_so_luong_sp: bien doi so luong san pham do nha quan tri de ra
    # d_chi_phi_co_dinh: bien doi chi phi co dinh do nha quan tri de ra
    doanh_thu_moi = gia_ban_sp * so_luong_sp_moi
    tong_bien_phi_moi = tong_bien_phi * so_luong_sp_moi / so_luong_sp
    tong_loi_nhuan_gop = doanh_thu - tong_bien_phi
    tong_loi_nhuan_gop_moi = doanh_thu_moi - tong_bien_phi_moi
    loi_nhuan = tong_loi_nhuan_gop - tong_chi_phi_co_dinh
    loi_nhuan_moi = tong_loi_nhuan_gop_moi - tong_chi_phi_co_dinh_moi
    d_loi_nhuan = loi_nhuan_moi - loi_nhuan

    # may cai input vao duoi dang list
    return d_loi_nhuan

# Thay doi bien phi va doanh thu

def optimize_bienphi(gia_ban_sp, doanh_thu, so_luong_sp, so_luong_sp_moi, tong_bien_phi, tong_bien_phi_moi):
    # d_so_luong_sp: bien doi so luong san pham do nha quan tri de ra
    # d_bien_phi: bien doi bien phi do nha quan tri de ra
    doanh_thu_moi = gia_ban_sp * so_luong_sp_moi
    tong_bien_phi_moi = tong_bien_phi * so_luong_sp_moi / so_luong_sp
    tong_loi_nhuan_gop = doanh_thu - tong_bien_phi
    tong_loi_nhuan_gop_moi = doanh_thu_moi - tong_bien_phi_moi
    d_loi_nhuan = tong_loi_nhuan_gop_moi - tong_loi_nhuan_gop
    
    return d_loi_nhuan

# Thay doi co cau san pham va doanh thu

def optimize_cocau_cuthe(so_luong_sp, gia_ban, bien_phi_per_sp, tong_chi_phi_co_dinh, d_doanh_thu:np.array, d_tong_san_pham:np.array):
    # san_pham, gia_ban, co_cau_tieu_thu, bien_phi_per_sp, d_doanh_thu, d_tong_san_pham: np.array[float]
    n = len(so_luong_sp)
    loi_nhuan_gop_per_sp = gia_ban - bien_phi_per_sp
    ty_le_loi_nhuan_gop = loi_nhuan_gop_per_sp / gia_ban

    tong_loi_nhuan_gop = np.sum(so_luong_sp * loi_nhuan_gop_per_sp)
    tong_bien_phi = np.sum(so_luong_sp * bien_phi_per_sp)
    doanh_thu = so_luong_sp * gia_ban
    loi_nhuan = tong_loi_nhuan_gop - tong_chi_phi_co_dinh

    d_tong_san_pham = d_tong_san_pham + d_doanh_thu / gia_ban
    so_luong_sp_moi = so_luong_sp + d_tong_san_pham
    co_cau_tieu_thu_moi = so_luong_sp_moi / np.sum(so_luong_sp_moi)
    tong_loi_nhuan_gop_moi = np.sum(so_luong_sp_moi * loi_nhuan_gop_per_sp)
    tong_bien_phi_moi = np.sum(so_luong_sp_moi * bien_phi_per_sp)
    d_loi_nhuan = tong_loi_nhuan_gop_moi - tong_chi_phi_co_dinh

    return d_loi_nhuan
```

```python
# Merge all functions into one

def cvp_analysis_all(so_luong_sp:np.array, gia_ban_sp:np.array, doanh_thu: float, bien_phi_per_sp:np.array, bien_phi_per_sp_moi:float,  tong_chi_phi_co_dinh:float, tong_chi_phi_co_dinh_moi:float, d_doanh_thu:np.array, d_so_luong_sp:np.array):
    # cac chi tieu cu
    tong_bien_phi = np.sum(so_luong_sp * bien_phi_per_sp)
    loi_nhuan_gop_per_sp = gia_ban_sp - bien_phi_per_sp
    tong_loi_nhuan_gop = np.sum(so_luong_sp * loi_nhuan_gop_per_sp)
    loi_nhuan_cu = tong_loi_nhuan_gop - tong_chi_phi_co_dinh
    co_cau_tieu_thu = so_luong_sp / np.sum(so_luong_sp)

    # cac chi tieu moi
    tong_bien_phi_moi = bien_phi_per_sp_moi * d_so_luong_sp
    so_luong_sp_moi = so_luong_sp + d_so_luong_sp * so_luong_sp + d_doanh_thu / gia_ban_sp
    co_cau_tieu_thu_moi = so_luong_sp_moi / np.sum(so_luong_sp_moi)
    loi_nhuan_gop_per_sp_moi = gia_ban_sp - bien_phi_per_sp_moi
    tong_loi_nhuan_gop_moi = np.sum(so_luong_sp_moi * loi_nhuan_gop_per_sp_moi)
    loi_nhuan_moi = tong_loi_nhuan_gop_moi - tong_chi_phi_co_dinh_moi

    d_loi_nhuan = loi_nhuan_moi - loi_nhuan_cu

    return d_loi_nhuan

```

```python
# Create a function enables the manager to compare each target situations

def cvp_analysis_overall(*situation):
    # situation: list of tuple
    # (so_luong_sp:np.array, gia_ban_sp:np.array, doanh_thu: float, bien_phi_per_sp:np.array, bien_phi_per_sp_moi:float,  tong_chi_phi_co_dinh:float, tong_chi_phi_co_dinh_moi:float, d_doanh_thu:np.array, d_so_luong_sp:np.array)
    loi_nhuan = []
    for i in range(len(situation)):
        loi_nhuan.append(cvp_analysis_all(*situation[i]))
    loi_nhuan = np.array(loi_nhuan)
    # loi_nhuan = loi_nhuan / loi_nhuan[0]
    # loi_nhuan = loi_nhuan - 1
    return loi_nhuan
```
